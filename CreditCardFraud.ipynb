```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
# loading the dataset
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
# Separating the data from analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
# Under-sampling
## Build a sample dataset containig similar distribution of legit & fraud transactions

legit_sample = legit.sample(n=492)
```

```python
legit_sample.shape
```

```python
# Concat two dataframes
# axis = 0 -> add one by one rows
new_df = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df.head()
```

```python
new_df['Class'].value_counts()
```

```python
# Now data is uniformely distributed
```

```python
new_df.groupby('Class').mean()
```

```python
# here we can understand is nature of the data remains similar even after
# we took only 492 rows. Just look the mean differences of V1 in credit_card_data and new_df

```

```python
# Splitting the data into features and targets
X = new_df.drop(columns='Class',axis=1)
Y = new_df['Class']
```

```python
X
```

```python
Y
```

```python
# Split the data into Training and Testing

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2, stratify=Y, random_state=2)

# stratify used to stratify based on Y
# to maintain distribution of 1 and 0 in training and testing data
```

```python
# Model Training using Logistic Regression for Binary CLassification model

model = LogisticRegression()
```

```python
# training the LR model with training data
model.fit(x_train,y_train)
# y_train has corresponding label for X_train feature
```

```python
# Model Evaluation on accuracy score on training data

x_train_pred = model.predict(x_train)
training_data_acc = accuracy_score(x_train_pred,y_train)

training_data_acc
```

```python
# accuracy score on Test data

x_test_pred = model.predict(x_test)
testing_data_acc = accuracy_score(x_test_pred,y_test)

testing_data_acc
```

```python
print(f"Accuracy score of training data = {training_data_acc}\nAccuracy score of Test data = {testing_data_acc}")
```

```python
# So why we did accuracy prediction on training data?
# If the training data accuracy came 50% then it means we overfit the data
# If we get low train data accuracy and very high test data accuracy then model is underfitted
```
